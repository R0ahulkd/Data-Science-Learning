```python
import matplotlib.pyplot as plt

y = [98,67,88,95,88]
x = ["Part 1","Part 2","Part 3","Part 4","Part 5"]

# This program creates a bar chart to display values for different parts using Matplotlib.
plt.bar(x,y)
plt.savefig("01_bar.png")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [45,34,56,23,45]
plt.subplot(2,2,1)
plt.plot(x,y)

x = [5,6,7,8,9]
y = [67,50,66,56,82]
plt.subplot(2,2,2)
plt.plot(x,y)

x = [2,4,6,8,10]
y = [57,50,60,55,20]
plt.subplot(2,2,3)
plt.plot(x,y)

x = [2,4,6,8,10]
y = [57,50,60,55,20]
plt.subplot(2,2,4)
plt.plot(x,y)
plt.savefig("02_Subplot.png")
plt.show()
```

```python
brands = ["Oneplus","Apple","Samsung","Nokia","Redmi"]
popularity = [22,35,30,3,10]
c = ["red","silver","blue","pink","green"]
ex = [0,0,0,0,0.1]

plt.pie(popularity, labels = brands, colors = c, explode = ex, shadow = True, autopct = "%.f")
plt.savefig("02_Piechart.png")
plt.show()
```
