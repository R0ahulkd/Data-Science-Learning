```python
x = ["day1","day2","day3","day4","day5"]
y = [300,420,250,230,400]
y1 = [500,450,300,250,320]

plt.plot(x,y, marker = "^", ls = "--", label = "Week 1")
plt.plot(x,y1, marker = "^", ls = "-", color = "brown", label = "Week 2", alpha = 0.5) # Alpha makes transparent
plt.legend()
plt.show()
```
