```python
import matplotlib.pyplot as plt
import pandas as pd

x = ["Day 1","Day 2","Day 3","Day 4","Day 5"]
y = [30,40,25,30,40]

plt.step(x,y)
plt.show()
```

```python
x = ["Day 1","Day 2","Day 3","Day 4","Day 5"]
y = [30,40,25,30,40]

plt.step(x,y, where = "post")
plt.show()
```

```python
x = ["Day 1","Day 2","Day 3","Day 4","Day 5"]
y = [30,40,25,30,40]

plt.step(x,y, where = "mid")
plt.show()
```

```python
x = ["Day 1","Day 2","Day 3","Day 4","Day 5"]
y = [30,40,25,30,40]

plt.step(x,y, where = "mid", marker = "o")
plt.show()
```
