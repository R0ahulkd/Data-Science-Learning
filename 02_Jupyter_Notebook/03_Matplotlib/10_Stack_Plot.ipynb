```python
import matplotlib.pyplot as plt
import pandas as pd

days = [1,2,3,4,5,6,7]

NOP1 = [5,10,30,20,35,60,80]
NOP2 = [10,20,30,25,40,50,90]
NOP3 = [8,30,50,60,70,90,100]

plt.stackplot(days, NOP1, NOP2, NOP3)
plt.show()
```

```python
NOP1 = [5,10,30,20,35,60,80]
NOP2 = [10,20,30,75,40,90,120]
NOP3 = [8,30,50,60,70,90,100]

plt.stackplot(days, NOP1, NOP2, NOP3)
plt.show()
```

```python
NOP1 = [5,10,30,20,35,60,80]
NOP2 = [50,60,30,75,80,90,120]
NOP3 = [8,30,50,60,70,90,100]

plt.stackplot(days, NOP1, NOP2, NOP3)
plt.show()
```

```python
NOP1 = [5,10,30,20,35,60,80]
NOP2 = [50,60,30,75,80,90,120]
NOP3 = [8,30,50,60,70,90,100]

plt.stackplot(days, NOP1, NOP2, NOP3, baseline = "sym")
plt.show()
```

```python
NOP1 = [5,10,30,20,35,60,80]
NOP2 = [50,60,30,75,80,90,120]
NOP3 = [8,30,50,60,70,90,100]

plt.stackplot(days, NOP1, NOP2, NOP3, baseline = "wiggle")
plt.show()
```

```python
NOP1 = [5,10,30,20,35,60,80]
NOP2 = [50,60,30,75,80,90,120]
NOP3 = [8,30,50,60,70,90,100]

plt.stackplot(days, NOP1, NOP2, NOP3, colors = ["pink","hotpink","magenta"])
plt.show()
```

```python
NOP1 = [5,10,30,20,35,60,80]
NOP2 = [50,60,30,75,80,90,120]
NOP3 = [8,30,50,60,70,90,100]

plt.stackplot(days, NOP1, NOP2, NOP3, colors = ["pink","hotpink","magenta"], labels = ["Week 1", "Week 2", "Week 3"])
plt.legend()
plt.show()
```
