```python
import matplotlib.pyplot as plt

y = [98,67,88,95,88]
x = ["Part 1","Part 2","Part 3","Part 4","Part 5"]

# This program creates a bar chart to display values for different parts using Matplotlib.
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y)
plt.xlabel("parts of Harry Potter")
plt.ylabel("Popularity")
plt.title("Popularity of Different parts of Harry Potter")
plt.show()
```

```python
plt.bar(x,y)
plt.xlabel("parts of Harry Potter", fontsize = 10)
plt.ylabel("Popularity", fontsize = 10)
plt.title("Popularity of Different parts of Harry Potter", fontsize = 15)
plt.show()
```

```python
plt.bar(x,y, color = "green")
plt.xlabel("parts of Harry Potter", fontsize = 10)
plt.ylabel("Popularity", fontsize = 10)
plt.title("Popularity of Different parts of Harry Potter", fontsize = 15)
plt.show()
```

```python
colors = ["red","green","yellow","blue","orange"]
plt.bar(x,y, color = colors, edgecolor = "black")
plt.xlabel("parts of Harry Potter", fontsize = 10)
plt.ylabel("Popularity", fontsize = 10)
plt.title("Popularity of Different parts of Harry Potter", fontsize = 15)
plt.show()
```
