```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

brands = ["Oneplus","Apple","Samsung","Nokia","Redmi"]
popularity = [22,35,30,3,10]

plt.pie(popularity)
plt.show()
```

```python
plt.pie(popularity, labels = brands)
plt.show()
```

```python
brands = ["Oneplus","Apple","Samsung","Nokia","Redmi"]
popularity = [22,35,30,3,10]
c = ["red","silver","blue","pink","green"]

plt.pie(popularity, labels = brands, colors = c)
plt.show()
```

```python
brands = ["Oneplus","Apple","Samsung","Nokia","Redmi"]
popularity = [22,35,30,3,10]
c = ["red","silver","blue","pink","green"]
ex = [0,0,0,0,0.1]

plt.pie(popularity, labels = brands, colors = c, explode = ex)
plt.show()
```
