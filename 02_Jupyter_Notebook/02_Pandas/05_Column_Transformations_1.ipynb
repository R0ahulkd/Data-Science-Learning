```python
import pandas as pd

data = pd.read_excel("Employee Sample Data.xlsx")
print(data)
```

```python
# Check Bonus %: mark employees with 0 as "no bonus" and >0 as "bonus"
# Used .loc to filter rows and assign text values to the GetBonus column

data.loc[(data["Bonus %"] == 0), "GetBonus"] = "no bonus"
data.loc[(data["Bonus %"] > 0, "GetBonus")] = "bonus"
print(data)
```

```python
(data['GetBonus'] == 'bonus').sum() #475 People recieves bonus
```

```python
(data['GetBonus'] == 'no bonus').sum()
```
