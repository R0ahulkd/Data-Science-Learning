```python
import pandas as pd

data1 = {"Emp ID":["E01","E02","E03","E04","E05","E06"],
        "Names":["Ram","Shyam","Rahul","Vishal","Ravi","Rajan"],
        "Age":[34,56,23,44,32,34]}

data2 = {"Emp ID":["E01","E02","E03","E04","E05","E06"],
        "Salary":[45000,56000,34000,30000,50000,42000]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
print(df1)
print()
print(df2)
```

```python
print(pd.merge(df1,df2,on = "Emp ID"))
# Merge df1 and df2 using Emp ID as the common key (inner join)
# Combines rows where Emp ID exists in both DataFrames
```

```python
import pandas as pd

data1 = {"Emp ID":["E01","E02","E03","E04","E05","E06"],
        "Names":["Ram","Shyam","Rahul","Vishal","Ravi","Rajan"],
        "Age":[34,56,23,44,32,34]}

data2 = {"Emp ID":["E01","E07","E03","E04","E08","E06"],
        "Salary":[45000,56000,34000,30000,50000,42000]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
print(df1)
print()
print(df2)
```

```python
print(pd.merge(df1,df2,on = "Emp ID")) # Now data change as we change some data
```

```python
print(pd.merge(df1,df2,on = "Emp ID", how = "left"))
# Same as print(pd.merge(left = df1,eight = df2,on = "Emp ID", how = "left"))
# Left join: keep all rows from df1 and match rows from df2 using Emp ID
# Missing matches from df2 become NaN
# Useful when df1 is the primary dataset and df2 adds extra info
```

```python
print(pd.merge(df1,df2,on = "Emp ID", how = "right"))
```
