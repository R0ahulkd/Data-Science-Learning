```python
import pandas as pd

data = pd.read_excel("Employee Sample Data.xlsx")
print(data)
```

```python
print(data.head(10))
```

```python
gp = data.groupby("Department").agg({"Gender":"count"})
print(gp)

# Group rows by Department and count how many entries each department has
# Using Gender column for counting non-null values
# Result shows total number of employees per department
```

```python
gp = data.groupby("Job Title").agg({"EEID":"count"})
print(gp)
```

```python
# Group by Department and Gender to count how many employees belong to each category
# Counts EEID values to get the total employees per Department-Gender combination
# Useful to analyze gender distribution inside departments


gp = data.groupby(["Department","Gender"]).agg({"EEID":"count"})
print(gp)
```

```python
gp1 = data.groupby("Country").agg({"Age":"mean"})
print(gp1)
```

```python
gp1 = data.groupby("Country").agg({"Annual Salary":"mean"})
print(gp1)
```

```python
gp1 = data.groupby("Country").agg({"Annual Salary":"max"})
print(gp1)
```

```python
gp1 = data.groupby(["Country","Gender"]).agg({"Annual Salary":"max"})
print(gp1)
```

```python
gp1 = data.groupby(["Country","Gender"]).agg({"Annual Salary":"min"})
print(gp1)
```

```python
gp1 = data.groupby(["Country","Gender"]).agg({"Annual Salary":"min"})
print(gp1)
```
