```python
import pandas as pd

dict = {"keys":["k1","k2","k1","k2"],
       "Name":["John","Ben","David","Peter"],
       "Houses":["red","blue","green","red"],
       "Grades":["3rd","8th","9th","8th"]}

df = pd.DataFrame(dict)
print(df)
```

```python
df.pivot(index="keys", columns="Name", values="Houses")
# Reshape data by using keys as rows, Names as columns, and Houses as values
```

```python
df.pivot(index="keys", columns="Name", values=["Houses","Grades"])
```

```python
dict = {"Name":["John","Ben","David","Peter"],
       "Houses":["red","blue","green","red"],
       "Grades":["3rd","8th","9th","8th"]}

df = pd.DataFrame(dict)
print(df)

print(pd.melt(df, id_vars=["Name"], value_vars = ["Houses"])) # Convert wide data to long format keeping Name as id and unpivoting Houses column
```

```python
print(pd.melt(df, id_vars=["Name"], value_vars = ["Houses","Grades"]))
```

```python
print(pd.melt(df, id_vars=["Name"], value_vars = ["Houses","Grades"], var_name = "Houses&Grades", value_name = "values"))
```
