```python
import numpy as np

a = np.array([20,40,60,70])

print(np.sum(a))
```

```python
print(np.max(a))
```

```python
print(np.min(a))
```

```python
print(np.size(a)) # Size of array 4 elements
```

```python
print(np.mean(a))
```

```python
print(np.cumsum(a)) # Cummulative Sum
```

```python
print(np.cumprod(a)) # Cummulative Product of a
```

```python
a = [100,150,199,200,250,130]
b = [10,50,30,40,30,10]

price = np.array(a)
quantity = np.array(b)
print(price, "\n" ,quantity)

print(np.sum(price))
```

```python
print(np.cumprod([price,quantity], axis = 0)) # It prints like 100*10 and 150*50
```

```python
total = np.cumprod([price,quantity], axis = 0)
print(total)
print(total[1])
```

```python
print(total[1].sum())
```
