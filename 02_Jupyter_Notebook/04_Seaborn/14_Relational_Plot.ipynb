```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("tips")
print(data)
```

```python
sns.relplot(data = data, x = "tip",y = "total_bill")
plt.show()
```

```python
sns.relplot(data = data, x = "tip",y = "total_bill", hue = "sex")
plt.show()
```

```python
sns.relplot(data = data, x = "tip",y = "total_bill", hue = "sex", kind = "line")
plt.show()
```

```python
sns.relplot(data = data, x = "tip",y = "total_bill", hue = "sex", col = "day")
plt.show()
```

```python
sns.relplot(data = data, x = "tip",y = "total_bill", hue = "sex", col = "day", size = "smoker")
plt.show()
```
