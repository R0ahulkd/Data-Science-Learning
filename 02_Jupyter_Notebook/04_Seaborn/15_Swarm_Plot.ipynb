```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = sns.load_dataset("tips")
print(data)
```

```python
sns.swarmplot(data = data, x = "day", y = "total_bill")
plt.show()
```

```python
sns.stripplot(data = data, x = "day", y = "total_bill")
plt.show()
```

```python
sns.stripplot(data = data, x = "day", y = "total_bill", jitter = 0.3)
plt.show()
```

```python
sns.swarmplot(data = data, x = "day", y = "total_bill", hue = "sex")
plt.show()
```

```python
sns.swarmplot(data = data, x = "day", y = "total_bill", hue = "sex", dodge = True)
plt.show()
```

```python
sns.stripplot(data = data, x = "day", y = "total_bill", jitter = 0.3, hue = "sex")
plt.show()
```

```python
sns.stripplot(data = data, x = "day", y = "total_bill", jitter = 0.3, hue = "sex", dodge = True)
plt.show()
```
