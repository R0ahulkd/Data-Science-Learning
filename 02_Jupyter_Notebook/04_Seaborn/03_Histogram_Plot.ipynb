```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")
print(data)
```

```python
sns.histplot(data, x = "day")
plt.show()
```

```python
sns.histplot(data, x = "day", hue = "sex")
plt.show()
```

```python
sns.histplot(data, x = "day", hue = "sex", kde = True)
plt.show()
```

```python
data = sns.load_dataset("titanic")
print(data)
```

```python
sns.histplot(data = data, x = "age")
plt.show()
```

```python
sns.histplot(data = data, x = "age", hue = "sex")
plt.show()
```

```python
sns.histplot(data = data, x = "age", hue = "who")
plt.show()
```

```python
sns.histplot(data = data, x = "age", hue = "who", kde = True)
plt.show()
```

```python
sns.histplot(data = data, x = "age", hue = "who", kde = True, discrete = True)
plt.show()
```

```python
sns.histplot(data = data, x = "age", hue = "who", kde = True, bins = 8)
plt.show()
```

```python
sns.histplot(data = data, x = "age", hue = "who", kde = True, bins = 30)
plt.show()
```
