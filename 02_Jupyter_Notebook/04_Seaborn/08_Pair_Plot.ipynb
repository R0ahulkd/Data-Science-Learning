```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("tips")
print(data)
```

```python
sns.pairplot(data)
plt.show()
```

```python
sns.pairplot(data, hue = "day")
plt.show()
```

```python
sns.pairplot(data, hue = "day", diag_kind = "bar")
plt.show()
```

```python
sns.pairplot(data, hue = "day", diag_kind = "plot")
plt.show()
```

```python
sns.pairplot(data, hue = "day", diag_kind = "hist")
plt.show()
```

```python
sns.pairplot(data, hue = "day", palette = "viridis")
plt.show()
```
