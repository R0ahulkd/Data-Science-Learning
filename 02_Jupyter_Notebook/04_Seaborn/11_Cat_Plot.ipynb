```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = sns.load_dataset("tips")
print(data)
#Catplot stands for categorical plot
```

```python
sns.catplot(data = data, x = "tip", y = "day")
plt.show()
```

```python
sns.catplot(data = data, x = "day", y = "tip")
plt.show()
```

```python
sns.catplot(data = data, x = "day", y = "tip", hue = "sex")
plt.show()
```

```python
sns.catplot(data = data, x = "day", y = "tip", hue = "sex", kind = "bar")
plt.show()
```

```python
sns.catplot(data = data, x = "day", y = "tip", hue = "sex", kind = "violin")
plt.show()
```

```python
sns.catplot(data = data, x = "day", hue = "sex", kind = "count")
plt.show()
```
