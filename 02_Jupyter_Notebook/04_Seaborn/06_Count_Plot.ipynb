```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("tips")
print(data)
```

```python
sns.countplot(data = data, x = "sex")
plt.show()
```

```python
sns.countplot(data = data, x = "sex", hue = "day")
plt.show()
```
