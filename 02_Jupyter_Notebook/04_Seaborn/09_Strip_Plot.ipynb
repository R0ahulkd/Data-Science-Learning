```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = sns.load_dataset("tips")
print(data)
```

```python
sns.stripplot(data = data, x = "day", y = "total_bill")
plt.show()
```

```python
sns.scatterplot(data = data, x = "day", y = "total_bill") # Scatter Plot
plt.show()
```

```python
sns.stripplot(data = data, x = "day", y = "total_bill", hue = "sex")
plt.show()
```

```python
sns.stripplot(data = data, x = "day", y = "total_bill", hue = "sex", dodge = True)
plt.show()
```

```python
sns.stripplot(data = data, x = "day", y = "total_bill", hue = "sex", dodge = True, jitter = 0.3)
plt.show()
```

```python
sns.stripplot(data = data, x = "day", y = "total_bill", hue = "sex", dodge = True, jitter = 0.4)
plt.show()
```
